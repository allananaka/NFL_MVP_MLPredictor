```python
import pandas as pd
mvps = pd.read_csv("mvps.csv")
mvps.tail()
```

```python
mvps = mvps[mvps['year'] < 2022]
mvps = mvps[["Player", "Votes", "year"]]
mvps.tail()
```

```python
players = pd.read_csv("./nfl_data/yearly_player_data.csv")
players.head()
```

```python
w = players.groupby(['player_name', 'season']).filter(lambda x: len(x) > 1)
w
```

```python
players = players.drop(index=[1055, 1056, 1504, 1505, 4762, 4763])
w = players.groupby(['player_name', 'season']).filter(lambda x: len(x) > 1)
w
```

```python
player = players.rename(columns={'player_name': 'Player', 'season': 'year'}, inplace=True)
players.head()
```

```python
combined = players.merge(mvps, how='outer', on=['Player', 'year'])
combined['Votes'] = combined['Votes'].fillna(0)
```

```python
combined.head()
```

```python
combined = combined.drop(columns=['player_id', 'team'])
```

```python
combined.head()
```

```python
combined = pd.get_dummies(combined, columns=['position'])
```

```python
combined.head()
```

```python
combined['position_RB'] = combined['position_RB'].astype(int)
combined['position_QB'] = combined['position_QB'].astype(int)
combined['position_WR'] = combined['position_WR'].astype(int)
combined = combined.drop(columns=['position_TE'])
```

```python
#combined_noname = combined.drop(columns=['Player'])
combined_noname = combined
```
