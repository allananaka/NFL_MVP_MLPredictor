```python
years = list(range(2000, 2024))
url_template = "https://www.pro-football-reference.com/awards/awards_{}.htm#voting_apmvp"
```

```python
import requests

for year in years:
    url = url_template.format(year)
    data = requests.get(url)

    with open("previousMVPdata/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
from io import StringIO

dfs = []
for year in years:
    with open("previousMVPdata/{}.html".format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    table = soup.find(id="voting_apmvp")
    mvp = pd.read_html(StringIO(str(table)))[0]
    mvp["year"] = year

    dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)
mvps.to_csv("mvps.csv")
```
